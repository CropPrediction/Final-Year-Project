# Importing Dataset

```python
cropreq=pd.read_csv('croprequirements.csv') # d
```

```python
cropreq.head()
```

```python
crop_names=cropreq['Crop'].unique()
crop_names
```

```python
group_crop=cropreq.groupby('Crop')
eachcrop=[]
for i in crop_names:
    eachcrop.append(group_crop.get_group(i))
    
```

```python
eachcrop[0]
```

```python
cropreqnew=cropreq['Crop'].astype('category').to_frame()
```

```python
cropreqnew=cropreqnew.join(cropreq['Suitability'].astype('category'))
```

```python
cropreqnew
```

```python
cropreqnew=cropreqnew.join(cropreq['Texture'])
cleaned = cropreqnew.Texture.str.split(',', expand=True).stack()
a=pd.get_dummies(cleaned,prefix='g').groupby(level=0).sum()
cropreqnew=cropreqnew.join(a)
```

```python
cropreqnew=cropreqnew.join(cropreq['Slope_code'])
cleaned = cropreqnew.Slope_code.str.split(',', expand=True).stack()
a=pd.get_dummies(cleaned,prefix='g').groupby(level=0).sum()
cropreqnew=cropreqnew.join(a)
```
